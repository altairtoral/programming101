```python
class Person:
    def __init__(self, name, birthday, gender, height, weight, hairColor):
        self.name=name
        self.birthday=birthday
        self.gender= gender
        self.height = height
        self.weight= weight
        self.hairColor=hairColor
        
    def getBirthdayDate(self):
        birthDate=datetime.strptime(self.birthday,"%d.%m.%Y")
        return birthDate
        #return self.birthday.strftime('%d.%m.%Y')
```

```python
tona=Person ("Tona", "07.08.1982","M", 1.72,75,"black")
print (tona.name)
print (tona.birthday)
print (tona.gender)
a=tona.getBirthdayDate()
retirementAge=67
Age=thisYear-a.year
print (tona.name, "you are",Age, "years old")
yearstoWork=retirementAge-Age
print (tona.name,"you have to work",yearstoWork,"years")
#print (tona.birthday())
```
